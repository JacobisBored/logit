## Sinc Function

Calculate the sinc function: sin(r)/r.  Use a Cartesian x,y square grid
and calculate ``r = sqrt(x**2+y**2)`` with 0 in the center of the grid.
Calculate the function for -15,15 for both x and y.

Note: You can use 'newaxis' to ensure that y is a column vector instead of a row vector.  Also you can use 'seterr(all='ignore')' to ignore errors that crop up from division by zero.  Try to plot this on a grid as a polar coordinate function that has been translated to Cartesian coordinates.

```python
%matplotlib inline
import numpy as np
from matplotlib.pyplot import imshow, show, plot

x = linspace(-15,15,101)[None, :]
y = x.reshape(x.size, 1)
seterr(all='ignore')
# Your code here
#x,y = -15,15
r = sqrt(x**2 + y**2)

sinc = np.sin(r)/r
sinc[50,50] = 1

imshow(sinc, extent=[-15,15,-15,15])
```
